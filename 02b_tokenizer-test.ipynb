```python
import os
os.environ["HF_HOME"] = "/home/jovyan/.cache/huggingface"
import warnings
warnings.filterwarnings("ignore", message="IProgress not found")
from evo2 import Evo2

model = Evo2("evo2_7b")

# --- Test individual characters ---
test_chars = ["A", "C", "G", "T", "N"]
print("=== Single-character tokenization ===")
for ch in test_chars:
    tokens = model.tokenizer.tokenize(ch)
    print(f"  '{ch}' -> {tokens}  (len={len(tokens)})")

# --- Test short known sequences ---
test_seqs = [
    ("pure_ACGT", "ACGTACGTACGT"),
    ("with_N", "ACGTNNNACGT"),
    ("all_N", "NNNNNNNNNN"),
    ("50bp_bin", "A" * 50),
    ("mixed_50bp", "ACGT" * 12 + "NN"),
]

print("\n=== Sequence-level checks ===")
all_pass = True
for label, seq in test_seqs:
    tokens = model.tokenizer.tokenize(seq)
    ok = len(tokens) == len(seq)
    all_pass &= ok
    status = "OK" if ok else "FAIL"
    print(f"  [{status}] {label}: len(seq)={len(seq)}, len(tokens)={len(tokens)}")
    if not ok:
        print(f"         tokens: {tokens[:20]}{'...' if len(tokens) > 20 else ''}")

import random
random.seed(42)
seq_10k = "".join(random.choices("ACGT", k=9900)) + "N" * 100  # 100 Ns at end
tokens_10k = model.tokenizer.tokenize(seq_10k)
ok_10k = len(tokens_10k) == len(seq_10k)
all_pass &= ok_10k
status = "OK" if ok_10k else "FAIL"
print(f"  [{status}] 10kb_with_Ns: len(seq)={len(seq_10k)}, len(tokens)={len(tokens_10k)}")

print(f"\n{'ALL CHECKS PASSED' if all_pass else 'SOME CHECKS FAILED — binning scheme needs revision'}")
```
