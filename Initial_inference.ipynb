```python
from evo2 import Evo2
import torch
```

```python
print("Loading Evo2 7B...")
model = Evo2('evo2_7b_base')
print("Model loaded.")
```

```python
sequence = "AGATCGATCGGGGCGGAGCTCGATCG"
```

```python
input_ids = torch.tensor(
    model.tokenizer.tokenize(sequence),
    dtype=torch.int,
).unsqueeze(0).to('cuda:0')
```

```python
outputs, _ = model(input_ids)
logits = outputs[0]

print('Logits: ', logits)
print('Shape (batch, length, vocab): ', logits.shape)
```

```python
logits_2d = outputs[0].squeeze(0)  # (26, 512)

predicted_tokens = torch.argmax(logits_2d, dim=-1)  # (26,) — one predicted token per position
print("Predicted token ids:", predicted_tokens)

# Decode predicted tokens back to nucleotides
# Remember: logits[i] predicts position i+1 (autoregressive shift)
# So predicted_tokens[i] is the model's best guess for position i+1
sequence = "AGATCGATCGGGGCGGAGCTCGATCG"

print("\nPosition-by-position comparison:")
print(f"{'Pos':<6}{'Actual (i+1)':<16}{'Predicted':<16}{'Match'}")
print("-" * 44)
for i in range(len(sequence) - 1):
    actual = sequence[i + 1]
    predicted_id = predicted_tokens[i].item()
    predicted_char = model.tokenizer.detokenize([predicted_id])
    match = "✓" if actual == predicted_char else "✗"
    print(f"{i+1:<6}{actual:<16}{predicted_char:<16}{match}")
```

```python
print(f"{'Pos':<6}{'Actual':<10}{'Correct token rank':<22}{'Logit of correct':<20}{'Logit of top-1'}")
print("-" * 65)

for i in range(len(sequence) - 1):
    actual_char = sequence[i + 1]
    actual_id = model.tokenizer.tokenize(actual_char)[0]
    
    position_logits = logits_2d[i]  # (512,)
    
    # rank of correct token (0 = best)
    sorted_ids = torch.argsort(position_logits, descending=True)
    rank = (sorted_ids == actual_id).nonzero(as_tuple=True)[0].item()
    
    correct_logit = position_logits[actual_id].item()
    top1_logit = position_logits[sorted_ids[0]].item()
    
    print(f"{i+1:<6}{actual_char:<10}{rank:<22}{correct_logit:<20.3f}{top1_logit:.3f}")
```

```python
i = 10  # position 11 (0-indexed), G ranked 3rd

position_logits = logits_2d[i]
sorted_ids = torch.argsort(position_logits, descending=True)

print("Top 5 predictions at position 11:")
print(f"{'Rank':<8}{'Token ID':<12}{'Token':<12}{'Logit'}")
print("-" * 40)
for rank in range(5):
    token_id = sorted_ids[rank].item()
    token_char = model.tokenizer.detokenize([token_id])
    logit = position_logits[token_id].item()
    print(f"{rank:<8}{token_id:<12}{repr(token_char):<12}{logit:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ============================================================
# Evo2 7B — Embedding extraction and inspection
# ============================================================

layer_name = 'blocks.28.mlp.l3'

with torch.no_grad():
    outputs_emb, embeddings = model(
        input_ids,
        return_embeddings=True,
        layer_names=[layer_name]
    )

# embeddings[layer_name] shape: (1, seq_len, hidden_dim)
emb = embeddings[layer_name][0]  # strip batch dim -> (seq_len, hidden_dim)
print(f"Raw embedding shape (seq_len, hidden_dim): {emb.shape}")

# move to cpu and convert to float32 for numpy
emb_np = emb.cpu().float().numpy()  # (26, 4096)

# ---- basic statistics ----
print(f"\nPer-dimension statistics (across all positions):")
print(f"  Mean:  {emb_np.mean():.6f}")
print(f"  Std:   {emb_np.std():.6f}")
print(f"  Min:   {emb_np.min():.6f}")
print(f"  Max:   {emb_np.max():.6f}")

# ---- mean pooling -> sequence-level vector ----
mean_emb = emb_np.mean(axis=0)  # (4096,)
print(f"\nMean-pooled embedding shape: {mean_emb.shape}")
print(f"  Mean:  {mean_emb.mean():.6f}")
print(f"  Std:   {mean_emb.std():.6f}")

# ---- visualise per-position embedding norms ----
# L2 norm of each position's embedding vector
# positions with high norm have stronger/more distinctive representations
position_norms = np.linalg.norm(emb_np, axis=1)  # (26,)
print(f"\nPer-position L2 norms:")
for i, (norm, nuc) in enumerate(zip(position_norms, sequence)):
    print(f"  pos {i:>2} ({nuc}): {norm:.3f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# plot 1: per-position norms
axes[0].bar(range(len(sequence)), position_norms, color='steelblue', alpha=0.85)
for i, (norm, nuc) in enumerate(zip(position_norms, sequence)):
    axes[0].text(i, norm + 0.5, nuc, ha='center', va='bottom', fontsize=8)
# highlight SP1 core motif
for x in range(9, 15):
    axes[0].axvspan(x - 0.5, x + 0.5, alpha=0.12, color='gold')
axes[0].set_xlabel('Sequence position')
axes[0].set_ylabel('L2 norm')
axes[0].set_title('Per-position embedding L2 norms')

# plot 2: heatmap of embedding values across positions
# subsample dimensions for visibility (4096 is too many to plot)
subsample = emb_np[:, :200]  # first 200 dimensions
im = axes[1].imshow(subsample.T, aspect='auto', cmap='RdBu_r')
axes[1].set_xlabel('Sequence position')
axes[1].set_ylabel('Embedding dimension (first 200)')
axes[1].set_title('Embedding values across positions')
plt.colorbar(im, ax=axes[1])

plt.tight_layout()
plt.savefig('evo2_sp1_embeddings.png', dpi=150)
plt.show()
```
