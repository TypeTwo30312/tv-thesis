```python
import os
os.environ["HF_HOME"] = "/home/jovyan/.cache/huggingface"

import torch
import numpy as np
from pathlib import Path
import polars as pl
from evo2 import Evo2

# Paths
DATA_ROOT = Path.home() / "vambersky_t/Data"
WINDOWS_DIR = DATA_ROOT / "extracted_windows_hg38"
EMBEDDINGS_DIR = DATA_ROOT / "embeddings_hg38"

# Target file — smallest, used for the revcomp diagnostic
TARGET = "ENCFF765CKK__MYC__GM12878"
TF = "MYC"

# Embedding config — must match 02_embeddings.ipynb exactly
LAYER = "blocks.28.mlp.l3"
BIN_SIZE = 50
N_BINS = 200
RANDOM_SEED = 42

model = Evo2("evo2_7b")
print("Model loaded.")
print(f"GPU: {torch.cuda.get_device_name(0)}, "
      f"{torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB total")
```

```python
RC_TABLE = str.maketrans("ACGTNacgtn", "TGCANtgcan")

def reverse_complement(seq: str) -> str:
    """Return the reverse complement of a DNA sequence.
    Non-ACGTN characters are preserved as-is via the translation table."""
    return seq.translate(RC_TABLE)[::-1]

def embed_sequence(seq: str) -> np.ndarray:
    """
    Tokenise, forward pass, extract LAYER activations,
    bin-pool to (N_BINS, 4096) fp16 numpy array.
    Identical to 02_embeddings.ipynb except sequence is pre-transformed by caller.
    """
    input_ids = torch.tensor(
        [model.tokenizer.eod_id] + model.tokenizer.tokenize(seq),
        dtype=torch.int,
    ).unsqueeze(0).to("cuda:0")

    with torch.no_grad():
        _, embeddings = model(input_ids, return_embeddings=True, layer_names=[LAYER])

    emb = embeddings[LAYER][:, 1:, :]  # drop BOS token
    n_tokens = emb.shape[1]
    assert n_tokens == BIN_SIZE * N_BINS, \
        f"Unexpected token count: {n_tokens}"

    binned = emb[0].reshape(N_BINS, BIN_SIZE, -1).mean(dim=1)  # (200, 4096)
    return binned.to(torch.float16).cpu().numpy()

# Smoke test — one row from the original CSV
csv_path = WINDOWS_DIR / TF / f"{TARGET}.full_peaks.csv.gz"
df = pl.read_csv(csv_path)
test_seq = df["sequence"][0]
test_rc = reverse_complement(test_seq)
assert len(test_rc) == len(test_seq), "RC length mismatch"
assert test_rc != test_seq, "RC identical to forward — check translation table"
out = embed_sequence(test_rc)
print(f"Smoke test passed. Output shape: {out.shape}, dtype: {out.dtype}")
```

```python
def get_output_paths(target: str, tf: str):
    """
    Separate output directory so revcomp embeddings never overwrite originals.
    Schema: embeddings/{TF}_revcomp/{TARGET}_revcomp.*
    """
    out_dir = EMBEDDINGS_DIR / f"{tf}_revcomp"
    out_dir.mkdir(parents=True, exist_ok=True)
    base = f"{target}_revcomp"
    npz_path = out_dir / f"{base}.embeddings.npz"
    parquet_path = out_dir / f"{base}.peak_ids.parquet"
    work_bin = Path("/home/jovyan/embedding_work") / f"{base}.work.bin"
    progress_txt = Path("/home/jovyan/embedding_work") / f"{base}.progress.txt"
    Path("/home/jovyan/embedding_work").mkdir(parents=True, exist_ok=True)
    return npz_path, parquet_path, work_bin, progress_txt

npz_path, parquet_path, work_bin, progress_txt = get_output_paths(TARGET, TF)
print(f"NPZ output:     {npz_path}")
print(f"Parquet output: {parquet_path}")
```

```python
CHECKPOINT_EVERY = 100

csv_path = WINDOWS_DIR / TF / f"{TARGET}.full_peaks.csv.gz"

# Skip if already complete
if npz_path.exists() and parquet_path.exists():
    print("Output already exists — skipping.")
else:
    df = pl.read_csv(csv_path)
    peak_ids = df["peak_id"].to_list()
    sequences = df["sequence"].to_list()
    n_peaks = len(sequences)
    print(f"Loaded {n_peaks} peaks from {csv_path.name}")

    # Confirm sequences are the expected length
    seq_lengths = set(len(s) for s in sequences)
    assert seq_lengths == {10_000}, f"Unexpected sequence lengths: {seq_lengths}"
    
    # Spot-check RC is self-inverse on first sequence
    assert reverse_complement(reverse_complement(sequences[0])) == sequences[0], \
        "RC is not self-inverse — check translation table"
    print(f"Pre-loop checks passed. {n_peaks} peaks, all 10,000 bp.")

    # Resume from checkpoint
    start_idx = 0
    if progress_txt.exists():
        start_idx = int(progress_txt.read_text().strip())
        print(f"Resuming from peak {start_idx}")

    # Initialise or reopen memmap
    fp = np.memmap(
        work_bin,
        dtype="float16",
        mode="r+" if start_idx > 0 else "w+",
        shape=(n_peaks, N_BINS, 4096),
    )

    import time
    t0 = time.time()

    for i in range(start_idx, n_peaks):
        rc_seq = reverse_complement(sequences[i])  # <-- only difference vs 02_embeddings
        fp[i] = embed_sequence(rc_seq)

        if (i + 1) % CHECKPOINT_EVERY == 0:
            fp.flush()
            progress_txt.write_text(str(i + 1))
            elapsed = time.time() - t0
            rate = (i + 1 - start_idx) / elapsed
            remaining = (n_peaks - i - 1) / rate / 3600
            print(f"  [{i+1}/{n_peaks}] {rate:.2f} peaks/s, "
                  f"~{remaining:.1f} h remaining")

    fp.flush()
    print("Inference complete. Saving final outputs...")

    # Save NPZ
    np.savez_compressed(npz_path, embeddings=np.array(fp))

    # Save parquet sidecar
    (pl.DataFrame({"peak_id": peak_ids})
       .write_parquet(parquet_path))

    # Clean up working files
    del fp
    work_bin.unlink(missing_ok=True)
    progress_txt.unlink(missing_ok=True)

    print(f"Done. NPZ: {npz_path}, parquet: {parquet_path}")

```
